```python
import pandas as pd
import numpy as np

# Sample data
data = {
    'day': ['2017-01-01', '2017-01-04', '2017-01-05', '2017-01-06', '2017-01-07', 
            '2017-01-08', '2017-01-09', '2017-01-10', '2017-01-11'],
    'temperature': [32.0, np.nan, 28.0, np.nan, 32.0, np.nan, np.nan, 34.0, 40.0],
    'windspeed': [6.0, 9.0, np.nan, 7.0, np.nan, np.nan, np.nan, 8.0, 12.0],
    'event': ['Rain', 'Sunny', 'Snow', np.nan, 'Rain', 'Sunny', np.nan, 'Cloudy', 'Sunny']
}

df1 = pd.DataFrame(data)



# Interpolating numeric columns
df2 = df1.copy()
df2[['temperature', 'windspeed']] = df2[['temperature', 'windspeed']].interpolate()

# Forward fill for non-numeric column
df2['event'] = df2['event'].ffill()

# Display the updated DataFrame
print(df2)
```
